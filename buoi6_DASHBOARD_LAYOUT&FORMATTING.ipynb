## 📚 PHẦN 1 — LÝ THUYẾT: NGUYÊN TẮC THIẾT KẾ DASHBOARD

---

### Dashboard tốt vs Dashboard kém

| Tiêu chí | Dashboard kém | Dashboard tốt |
|---|---|---|
| **Số lượng màu** | 6–8 màu khác nhau, tự phát | 2–3 màu chủ đạo, nhất quán |
| **Font chữ** | Nhiều font, nhiều size tùy ý | 1 font family, size theo cấp bậc rõ ràng |
| **Layout** | Charts và ô xếp ngẫu nhiên | Grid layout — thẳng hàng, cân đối |
| **Thông tin** | Hiển thị mọi thứ có thể | Chỉ hiển thị những gì cần thiết |
| **Cấp bậc thông tin** | Tất cả thứ đều nổi bật như nhau | KPI quan trọng nhất nổi bật nhất |
| **Gridlines & Headings** | Để mặc định | Đã tắt — dashboard thoáng hơn nhiều |
| **Người xem** | Phải tìm thông tin | Thông tin tự tìm đến mắt người xem |

---

### Nguyên tắc Z-Pattern — Mắt người đọc di chuyển thế nào?

Mắt người khi nhìn vào một trang/màn hình thường di chuyển theo hình chữ Z:

```
Góc trái trên  ──────────────────→  Góc phải trên
     ↓                                     ↙
     ↓                               (chéo xuống)
     ↓                              ↙
Góc trái giữa  ──────────────────→  Góc phải giữa
```

**Ứng dụng vào Dashboard:**
- **Góc trái trên:** Tiêu đề Dashboard — cần nhìn thấy ngay
- **Hàng ngang đầu tiên:** 4 KPI Cards — thông tin tóm tắt quan trọng nhất
- **Hàng giữa:** Slicers — điều khiển filter
- **Phần dưới:** Charts chi tiết — người xem đã được định hướng bởi KPI

---

### 5 Nguyên tắc thiết kế cốt lõi

**1. White Space (khoảng trắng) là bạn của bạn**
Charts sát nhau, không có khoảng cách = khó đọc. Để khoảng trống giữa các elements ít nhất 8–10px.

**2. Alignment — thẳng hàng tuyệt đối**
Mắt người ngay lập tức nhận ra khi một chart lệch 2mm so với chart bên cạnh. Dùng Align Tools thay vì căn bằng tay.

**3. Consistency — nhất quán từ đầu đến cuối**
Nếu KPI Card 1 dùng font 28, KPI Card 2 phải dùng font 28. Nếu Chart 1 không có border, Chart 2 cũng không có border.

**4. Hierarchy — cấp bậc thông tin rõ ràng**
Thứ tự ưu tiên: KPI Cards (to, đậm, màu nổi) > Chart Titles (vừa) > Axis Labels (nhỏ, nhạt).

**5. Less is More — bớt đi để tốt hơn**
Xóa legend thừa, xóa gridlines thừa, xóa border thừa, xóa data labels thừa (khi chart đã đủ rõ).

### Bước 1 — Di chuyển tất cả Charts sang Dashboard

Nếu Charts vẫn còn trên các sheet PivotTable, cần chuyển sang sheet Dashboard:

**Phương pháp A — Move Chart (khuyến nghị, giữ nguyên PivotChart connection):**
1. Click vào Chart → Tab **Chart Design** → **Move Chart**
2. Chọn **Object in:** → chọn `Dashboard` → OK
3. Chart chuyển sang sheet Dashboard, vẫn giữ kết nối với PivotTable gốc

**Phương pháp B — Cut & Paste:**
1. Click vào viền chart → **Ctrl+X**
2. Chuyển sang sheet `Dashboard` → **Ctrl+V**
3. Chart paste ở vị trí góc trên trái — cần di chuyển tiếp

---

### Bước 2 — Resize Charts đều nhau

Để 4 charts có size đồng nhất:

1. Ctrl+Click lần lượt 4 charts để chọn tất cả
2. Tab **Format** (Shape Format) xuất hiện → nhóm **Size**
3. Nhập kích thước cụ thể:
   - **Width:** 13cm (cho charts 2 cột)
   - **Height:** 8cm (cho tất cả charts)
4. Nhấn **Enter** — tất cả 4 charts cùng lúc đổi về kích thước đồng nhất

> ⚠️ **Lưu ý:** Khi chọn nhiều charts, thay đổi Size sẽ áp dụng cho tất cả. Nhưng thay đổi màu sắc hay format sẽ áp dụng cho chart cuối cùng được click — cần format từng chart riêng.

---

### Bước 3 — Căn chỉnh và phân phối đều (Align Tools)

Đây là bước tạo sự khác biệt giữa Dashboard nghiệp dư và chuyên nghiệp:

**Căn chỉnh 2 charts hàng trên:**
1. Ctrl+Click chọn **Chart 1** (Line) và **Chart 2** (Bar)
2. Tab **Format** → **Align** → **Align Top** (căn đỉnh bằng nhau)
3. **Align** → **Distribute Horizontally** (phân phối khoảng cách đều nhau)

**Căn chỉnh 2 charts hàng dưới:**
1. Ctrl+Click chọn **Chart 3** (Doughnut) và **Chart 4** (Column)
2. **Align** → **Align Top**
3. **Align** → **Distribute Horizontally**

**Căn chỉnh cột trái thẳng đứng:**
1. Ctrl+Click **Chart 1** và **Chart 3**
2. **Align** → **Align Left** (căn mép trái bằng nhau)

**Căn chỉnh cột phải thẳng đứng:**
1. Ctrl+Click **Chart 2** và **Chart 4**
2. **Align** → **Align Left**

---

### Bước 4 — Định vị Slicers

Slicers đặt giữa KPI Cards và Charts:

1. Click **Slicer Region** → di chuyển vào vùng hàng 8–9, cột B
2. Click **Slicer Category** → đặt cạnh Slicer Region
3. Click **Slicer Quarter** → đặt phía phải, trải ngang

**Căn chỉnh 3 Slicers cùng hàng:**
1. Ctrl+Click cả 3 Slicers
2. **Format** → **Align** → **Align Top** → **Distribute Horizontally**

---

### Bước 5 — Di chuyển chính xác bằng phím mũi tên

Khi chart hoặc Slicer gần đúng vị trí nhưng cần tinh chỉnh:
- Click vào object → dùng phím **mũi tên** trên bàn phím để di chuyển từng pixel
- Giữ **Alt** khi kéo → object tự bám vào lưới ô (snap to grid) — tiện để căn thẳng

---

### Bước 6 — Khóa vị trí objects (tùy chọn)

Sau khi bố cục xong, khóa để không vô tình xê dịch:
1. Ctrl+A → chọn tất cả objects trên sheet
2. Right-click → **Format Object** → tab **Properties**
3. Chọn **"Don't move or size with cells"**

1. Right-click vào vùng trống trong chart → **Format Chart Area**

3. **Border** → **Solid Line** → Color `#E2E8F0` → Width: 0.5pt

### Áp dụng Typography — Workflow

**Đổi font trên Charts:**
1. Click vào Chart
2. Right-click vào **Chart Title** → **Format Chart Title** → Font → Segoe UI, 12, Bold
3. Click vào **Axis Labels** → Font → Segoe UI, 9

> ⚠️ Excel không có "Segoe UI Semibold" như PowerPoint. Thay thế bằng **Segoe UI Bold** cho chart titles, hoặc dùng **Calibri Bold** nếu Segoe UI không có sẵn.

**Đồng bộ font tất cả Charts:**
Tiếc là Excel không có tính năng "Apply font to all charts at once". Cần format từng chart — mỗi chart mất khoảng 1–2 phút.

**Mẹo tiết kiệm thời gian:** Dùng **Format Painter** (cọ sơn) để sao chép format từ một object sang object khác:
1. Click vào text đã format đúng
2. Tab **Home** → **Format Painter** (biểu tượng cọ sơn)
3. Click vào text cần format → format được sao chép ngay lập tức

## 🚦 PHẦN 5 — CONDITIONAL FORMATTING CHO KPI CARDS

---

### Conditional Formatting là gì?

Conditional Formatting (CF) tự động **thay đổi màu ô hoặc thêm icon** dựa trên giá trị — không cần can thiệp thủ công. Trong Dashboard, CF giúp người xem nhận ra ngay KPI nào đang tốt / trung bình / xấu mà không cần đọc số.

---

### Ứng dụng 1 — Icon Sets cho Profit Margin

Áp dụng 3-mũi tên màu cho ô hiển thị Profit_Margin:

| Bước | Thao tác |
|---|---|
| **1** | Click vào ô KPI Card Lợi Nhuận (ô chứa công thức GETPIVOTDATA hoặc Profit Margin) |
| **2** | Tab **Home** → **Conditional Formatting** → **Icon Sets** → **3 Arrows (Colored)** |
| **3** | Hệ thống tự chọn ngưỡng — cần tinh chỉnh: click **Conditional Formatting** → **Manage Rules** |
| **4** | Click **Edit Rule** |
| **5** | Đổi **Type** từ "Percent" sang **"Number"** |
| **6** | Nhập ngưỡng: **Mũi tên xanh** (▲): `>= 0.25` | **Mũi tên vàng** (→): `>= 0.15` | **Mũi tên đỏ** (▼): `< 0.15` |
| **7** | OK → OK |

**Kết quả:**
- Profit Margin ≥ 25% → ▲ (xanh lá)
- Profit Margin 15%–25% → → (vàng)
- Profit Margin < 15% → ▼ (đỏ)

**Kết quả mẫu**

---

### Ứng dụng 2 — Màu nền KPI Card theo giá trị

Tô nền KPI Card Tổng Doanh Thu theo mức độ:

| Bước | Thao tác |
|---|---|
| **1** | Chọn ô KPI Value Doanh Thu |
| **2** | **Conditional Formatting** → **New Rule** |
| **3** | Chọn **"Format only cells that contain"** |
| **4** | Cài đặt rule: `Cell Value` → `greater than` → nhập ngưỡng doanh thu mục tiêu (ví dụ: `10000000000` = 10 tỷ) |
| **5** | Click **Format** → tab **Fill** → màu xanh nhạt `#D1FAE5` → OK |

### 📖 BUỔI 2 — CHUẨN BỊ DỮ LIỆU

### 📖 BUỔI 5 — SLICERS & KPI CARDS

**Kiến thức cốt lõi:**

**SLICERS:**

- 3 Slicers: Region (1 cột), Category (2 cột), Quarter (4 cột)
- **Report Connections** = bước quan trọng nhất: Right-click Slicer → Report Connections → tick tất cả PivotTables
- Multi-select: giữ Ctrl hoặc click nút Multi-Select trên Slicer
- Timeline Slicer: Insert Timeline → chọn Order_Date → lọc theo khoảng ngày

**KPI CARDS:**
- `GETPIVOTDATA("Revenue", PT_Region!$A$3)` → lấy Grand Total Revenue từ PivotTable, tự cập nhật theo Slicer
- `COUNTA(SalesData[Order_ID])` → đếm số đơn hàng (dùng COUNTA vì Order_ID là text)
- `AVERAGE(SalesData[Revenue])` → doanh thu trung bình mỗi đơn
- Format KPI: `#,##0,,"Tỷ"` (tỷ VND), `#,##0,"K"` (nghìn VND), `#,##0` (số nguyên)
- Cấu trúc KPI Card: Hàng Label (font 10, xám, UPPERCASE) + Hàng Value (font 28, đậm, tối)

**Từ khóa cần nhớ:** Report Connections, GETPIVOTDATA, COUNTA, Merge Across, #,##0,,
