## 📚 PHẦN 1 — LÝ THUYẾT: CHỌN ĐÚNG LOẠI CHART

---

### Tại sao chọn đúng loại chart lại quan trọng?

Một biểu đồ sai loại không chỉ trông xấu — nó còn **truyền đạt thông tin sai**. Ví dụ:
- Dùng Pie Chart để so sánh xu hướng 12 tháng → người đọc không hiểu gì
- Dùng Line Chart để so sánh 5 danh mục → nhìn như mớ dây điện chằng chịt

Nguyên tắc cốt lõi: **chọn chart theo câu hỏi bạn muốn trả lời**, không phải theo sở thích cá nhân.

---

### Bộ 4 chart cơ bản cho Sales Dashboard

| Loại Chart | Trả lời câu hỏi | Ví dụ ứng dụng |
|---|---|---|
| **Line Chart** | Biến động theo thời gian như thế nào? | Revenue theo 12 tháng, tăng trưởng YoY |
| **Bar Chart** (nằm ngang) | So sánh các nhóm với nhau — ai lớn hơn ai? | Doanh thu theo vùng miền, top cities |
| **Column Chart** (đứng) | So sánh có thứ tự (ranking) — top N | Top 10 sản phẩm, top 5 sales rep |
| **Doughnut / Pie** | Tỷ lệ, phần trăm đóng góp — cái gì chiếm bao nhiêu? | Category mix, Segment breakdown |

---

### Bảng mở rộng — Thêm chart nâng cao

| Loại Chart | Khi nào dùng | Chú ý |
|---|---|---|
| **Clustered Bar** | So sánh nhiều series cùng chiều | Tối đa 3–4 series để không rối |
| **Stacked Column** | Cả tổng lẫn cấu trúc thành phần | Khó đọc nếu > 4 màu |
| **Combo Chart** (Line + Bar) | 2 metrics khác đơn vị trên cùng chart | Revenue (bar) + Profit Margin (line) |
| **Scatter Plot** | Tương quan giữa 2 biến số | Discount vs Profit Margin |

---

### Progression theo buổi học

| Cấp độ | Chart types | Buổi áp dụng |
|---|---|---|
| 🟢 Cơ bản | Line, Bar, Column, Doughnut | **Buổi 4** |
| 🟡 Trung cấp | Clustered Bar, Stacked Column | **Buổi 4** (mở rộng) |
| 🔵 Nâng cao | Combo Chart (dual-axis), sparklines | Tự học sau Buổi 4 |
| 🔴 Chuyên nghiệp | Dynamic charts qua OFFSET, Named Ranges | Kỹ năng Excel nâng cao |

---

### Quy tắc vàng chọn màu sắc Dashboard

Một trong những lỗi phổ biến nhất của người mới là dùng quá nhiều màu. Nguyên tắc:

**"1 màu chủ đạo — 1 màu nhấn — grey cho phần còn lại"**

| Vai trò màu | Dùng cho | Ví dụ |
|---|---|---|
| **Primary color** | Bars/Columns chính, đường Line | Blue `#2563EB` |
| **Accent color** | Giá trị nổi bật (max, highlight) | Orange `#F59E0B` |
| **Neutral** | Các bars còn lại, gridlines | Grey `#9CA3AF` |
| **Background** | Chart area, plot area | White hoặc Light grey `#F9FAFB` |

## 🔬 PHẦN 2 — GIẢI PHẪU PIVOTCHART

---

### PivotChart khác Chart thường như thế nào?

| Đặc điểm | Chart thường | PivotChart |
|---|---|---|
| **Nguồn dữ liệu** | Range cố định (A1:C10) | PivotTable — tự động co giãn |
| **Khi lọc dữ liệu** | Không thay đổi | Tự cập nhật theo filter |
| **Kết nối Slicer** | Không hỗ trợ | ✅ Kết nối được — dùng ở Buổi 5 |
| **Khi Refresh PivotTable** | Không ảnh hưởng | Chart tự cập nhật theo |
| **Cách tạo** | Insert → Chart | Click PivotTable → Insert → Chart |

> 💡 **Kết luận:** Luôn dùng PivotChart (không phải Chart thường) khi làm Dashboard — vì khi Slicer lọc dữ liệu, tất cả PivotCharts sẽ tự động thay đổi theo.

---

### Các thành phần của một PivotChart

```
┌─────────────────────────────────────────────┐
│  [Chart Title]                              │  ← Tiêu đề chart
│                                             │
│  Y-Axis │ 500M ──────────────────────────   │  ← Trục dọc (Values)
│  Label  │ 400M ─────────────●──────────     │
│         │ 300M ──●──────────────────────    │
│         │ 200M ──────────────────────────   │
│         │ 100M ──────────────────────────   │ 
│         └──────────────────────────────     │
│           Jan  Feb  Mar  Apr  ...  Dec      │  ← Trục ngang (Categories)
│                                             │
│  [Legend: ● Revenue  ■ Profit]              │  ← Chú thích
│  [Field Buttons: Month ▼]                   │  ← Nút Filter của PivotChart
└─────────────────────────────────────────────┘
```

---

### Các thành phần có thể format

| Thành phần | Cách chọn | Format phổ biến |
|---|---|---|
| **Chart Title** | Click 1 lần vào title | Đổi font, cỡ chữ, màu |
| **Plot Area** | Click vào vùng vẽ (bên trong) | Đổi màu nền |
| **Data Series** | Click vào bar/line/slice | Đổi màu series |
| **Axes** | Click vào trục số / trục text | Đổi font, format số |
| **Data Labels** | Click vào nhãn số trên chart | Đổi vị trí, format số |
| **Legend** | Click vào chú thích | Dời vị trí, ẩn/hiện |
| **Field Buttons** | Right-click → Hide All Field Buttons | Ẩn đi trước khi paste vào Dashboard |

> ⚠️ **Lưu ý quan trọng:** Khi di chuyển PivotChart sang sheet Dashboard, hãy **ẩn Field Buttons** trước (right-click lên button → Hide All Field Buttons). Chúng chỉ cần thiết khi đang chỉnh sửa, không nên xuất hiện trên Dashboard cuối cùng.

---

### Workflow chuẩn tạo 1 PivotChart

```
1. Click vào PivotTable nguồn
      ↓
2. Insert → Chart → chọn loại
      ↓
3. Đặt Chart Title
      ↓
4. Format Axes (format số VND, %)
      ↓
5. Thêm/tắt Data Labels
      ↓
6. Chỉnh màu sắc (Chart Design → Change Colors)
      ↓
7. Xóa Legend nếu không cần
      ↓
8. Ẩn Field Buttons
      ↓
9. Move Chart → Sheet "Dashboard"
```

## 📈 CHART 1 — DOANH THU THEO THÁNG (Line Chart)

---

### Câu hỏi business cần trả lời

> **"Doanh thu biến động như thế nào qua 12 tháng? Tháng nào cao điểm, tháng nào thấp điểm?"**

Line Chart là lựa chọn tốt nhất khi cần thể hiện **xu hướng theo thời gian** — mắt người sẽ tự nhiên theo dõi đường để nhận ra pattern tăng/giảm/mùa vụ.

---

### Cấu hình chart

| Thuộc tính | Giá trị |
|---|---|

| **Trục X** | Tháng (Jan – Dec) |
| **Trục Y** | Revenue (đơn vị: triệu VND) |

## 📊 CHART 2 — DOANH THU THEO VÙNG MIỀN (Bar Chart)

---

### Câu hỏi business cần trả lời

> **"Vùng miền nào đang dẫn đầu doanh thu? Khoảng cách giữa vùng cao nhất và thấp nhất là bao nhiêu?"**

Bar Chart nằm ngang (Horizontal Bar) lý tưởng khi so sánh **ít hơn 10 danh mục có tên dài** — tên danh mục nằm ngang dễ đọc hơn xoay chéo.

---

### Cấu hình chart

| Thuộc tính | Giá trị |
|---|---|
| **Nguồn PivotTable** | `PT_Region` |
| **Loại chart** | Clustered Bar (nằm ngang) |
| **Tiêu đề** | Doanh thu theo vùng miền |
| **Trục X (ngang)** | Revenue (đơn vị: triệu VND) |
| **Trục Y (dọc)** | Tên 5 vùng miền |

---

### Các bước tạo chi tiết

| Bước | Thao tác |
|---|---|
| **1** | Chuyển sang sheet `PT_Region` → Click vào bất kỳ ô nào trong PivotTable |
| **2** | Tab **Insert** → **Bar Chart** → chọn **Clustered Bar** (hàng đầu tiên, loại nằm ngang) |
| **3** | Click vào **Chart Title** → gõ: `Doanh thu theo vùng miền` |
| **4** | Click vào **trục X** (số bên dưới) → Right-click → **Format Axis** → Number → Custom → `#,##0,,"M"` |
| **5** | Sort bars từ lớn đến nhỏ: Click vào bất kỳ bar nào → Right-click → **Sort** → **Sort Largest to Smallest** |

---

### Tại sao phải Sort bars?

Bars không được sort nhìn rất nghiệp dư. Khi sort từ lớn đến nhỏ:
- Mắt người đọc ngay lập tức thấy ranking
- Vùng dẫn đầu nằm trên cùng, vùng yếu nhất nằm dưới
- Không cần data labels cũng đọc được thứ hạng

---

### Format màu sắc Gradient Blue

**Cách 1 — Dùng preset của Excel:**
1. Click vào bất kỳ bar nào → chọn tất cả bars
2. Tab **Chart Design** → **Change Colors** → chọn palette có tone xanh dương

**Cách 2 — Tô màu thủ công từng bar (chuyên nghiệp hơn):**
1. Click chọn tất cả bars → Right-click → **Format Data Series** → **Fill** → **Solid Fill** → chọn màu xanh đậm
2. Click chậm vào riêng bar thấp nhất → Right-click → **Format Data Point** → **Fill** → màu xanh nhạt hơn
3. Hiệu ứng gradient tự nhiên: bar cao = màu đậm, bar thấp = màu nhạt → trực quan nhất định thứ hạng

## 🍩 CHART 3 — PHÂN BỔ THEO DANH MỤC (Doughnut Chart)

---

### Câu hỏi business cần trả lời

> **"Mỗi category chiếm bao nhiêu % tổng doanh thu? Cơ cấu danh mục có đang cân bằng không?"**

Doughnut Chart (và Pie Chart) dùng để thể hiện **tỷ lệ phần trăm**. Doughnut được ưa hơn Pie vì lỗ ở giữa có thể chứa KPI số tổng — thiết kế chuyên nghiệp hơn.

---

### Khi nào KHÔNG dùng Pie/Doughnut?

- Khi có hơn 5–6 lát cắt → khó phân biệt màu, các lát nhỏ gần giống nhau
- Khi cần so sánh chính xác giữa các phần → Bar Chart chính xác hơn
- Khi cần thể hiện thay đổi theo thời gian → dùng Line/Column thay thế

---

### Cấu hình chart

| Thuộc tính | Giá trị |
|---|---|
| **Nguồn PivotTable** | `PT_Category` |
| **Loại chart** | Doughnut |
| **Tiêu đề** | Phân bổ theo danh mục |
| **Labels** | Category Name + Percentage |

---

### Các bước tạo chi tiết

| Bước | Thao tác |
|---|---|
| **1** | Chuyển sang sheet `PT_Category` |
| **2** | **Quan trọng:** Trước khi tạo chart, thu gọn PivotTable chỉ còn tầng Category (click **[-]** để ẩn Sub_Category) → Chart sẽ chỉ vẽ 4 lát cắt lớn thay vì hàng chục Sub_Category |
| **3** | Click vào PivotTable → Insert → **Pie Chart** → **Doughnut** |
| **4** | Click vào **Chart Title** → gõ: `Phân bổ theo danh mục` |

---

### Thêm Data Labels đúng chuẩn

| Bước | Thao tác |
|---|---|
| **1** | Click vào chart → Click dấu **[+]** → **Data Labels** → **More Data Label Options...** |
| **2** | Trong Format Data Labels panel bên phải: tick **Category Name** |
| **3** | Tick **Percentage** |
| **4** | Bỏ tick **Value** (không cần hiển thị số tuyệt đối trong Doughnut) |
| **5** | **Separator:** chọn **New Line** để tên và % hiển thị trên 2 dòng |
| **6** | **Label Position:** chọn **Outside End** cho label dễ đọc hơn |

## 🏆 CHART 4 — TOP 10 SẢN PHẨM BÁN CHẠY (Column Chart)

---

### Câu hỏi business cần trả lời

> **"Sản phẩm nào đang tạo ra nhiều doanh thu nhất? Top 10 chiếm bao nhiêu % tổng doanh thu?"**

Column Chart (đứng) dùng tốt khi: (1) cần ranking, (2) nhãn category ngắn, (3) số lượng danh mục từ 5–12.

---

### Cấu hình chart

| Thuộc tính | Giá trị |
|---|---|
| **Nguồn PivotTable** | `PT_Products` (đã lọc Top 10) |
| **Loại chart** | Clustered Column |
| **Tiêu đề** | Top 10 sản phẩm bán chạy |
| **Trục X** | Tên sản phẩm |
| **Trục Y** | Revenue (VND) |

---

### Kiểm tra PivotTable trước khi tạo chart

Xác nhận `PT_Products` đang hiển thị đúng Top 10:
1. Mở sheet `PT_Products`
2. Nhìn vào Row Labels — phải thấy đúng 10 tên sản phẩm (nếu chưa lọc, xem lại Buổi 3 phần Value Filters)
3. Các sản phẩm phải đã được sort từ cao xuống thấp theo Revenue

---

### Các bước tạo chi tiết

| Bước | Thao tác |
|---|---|
| **1** | Sheet `PT_Products` → Click vào bất kỳ ô nào trong PivotTable |

### 📊 Chart 5 — Combo Chart: Revenue vs Profit Margin (PT_Profit)

**Câu hỏi:** "Tháng nào bán nhiều nhưng margin thấp (hy sinh lợi nhuận để chạy khuyến mãi)?"

Combo Chart (hay Dual-Axis Chart) là chart chuyên nghiệp nhất vì vẽ **2 metrics khác đơn vị** trên cùng 1 chart — Revenue (VND) trên trục trái, Profit Margin (%) trên trục phải.

| Thao tác | Chi tiết |
|---|---|
| **Nguồn** | `PT_Profit` — kéo `Category` vào Rows, `Revenue` và `Avg Profit_Margin` vào Values |
| **Tạo chart** | Click PivotTable → Insert → **Recommended Charts** → tab **All Charts** → **Combo** |
| **Cài đặt** | Revenue → **Clustered Column** (Primary Axis); Profit_Margin → **Line** (Secondary Axis) |
| **Tiêu đề** | `Doanh thu & Biên lợi nhuận theo danh mục` |
| **Trục trái** | Format: `#,##0,,"M"` |
| **Trục phải** | Format: `0%` |

**Cách tạo Combo Chart từ đầu:**
1. Click PivotTable → Insert → nhóm Charts → **Combo** (biểu tượng cuối cùng, bar + line kết hợp)
2. Hoặc: Tạo Column Chart trước → Tab **Chart Design** → **Change Chart Type** → **Combo**
3. Cài Revenue = Clustered Column; Profit_Margin = Line with Markers; tick **Secondary Axis** cho Profit_Margin

> 🎯 **Insight từ chart này:** Nếu Category nào có bar Revenue cao nhưng đường Margin thấp → đang bán nhiều nhưng không có lãi. Đây là cảnh báo quan trọng cho ban lãnh đạo.

**Kết quả mẫu**

### Xóa các thành phần làm rối chart

Nhiều thành phần Excel thêm mặc định thực ra làm chart **kém chuyên nghiệp hơn**:

Excel có sẵn bộ Chart Styles và Color palettes:

### Căn chỉnh size chart đồng đều trên Dashboard

Khi nhiều charts trên cùng 1 sheet, size không đều trông rất nghiệp dư. Cách căn chỉnh:

1. Click chart 1 → giữ **Ctrl** → click chart 2, 3, 4 (chọn nhiều cùng lúc)
2. Tab **Format** (xuất hiện khi chọn nhiều object) → **Size** → nhập Width và Height cụ thể
3. Ví dụ: tất cả charts nhỏ = **8cm × 8cm**; chart lớn (Line Chart) = **16cm × 8cm**
