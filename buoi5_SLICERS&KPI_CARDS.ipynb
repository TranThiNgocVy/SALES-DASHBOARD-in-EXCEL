## 📚 PHẦN 1 — LÝ THUYẾT: SLICER LÀ GÌ?

---

### So sánh: Slicer vs Dropdown Filter thông thường

| Tiêu chí | Dropdown Filter (cũ) | Slicer (mới) |
|---|---|---|
| **Giao diện** | Dropdown nhỏ ở góc ô | Bảng nút bấm rõ ràng, có thể đặt bất kỳ đâu |
| **Xem trạng thái lọc** | Phải click vào mới thấy | Nút sáng/tối ngay trên bảng |
| **Tốc độ thao tác** | Click dropdown → scroll → chọn | 1 click thẳng vào giá trị |
| **Đa chọn** | Ctrl+click hoặc tick từng ô | Click nhiều nút liên tiếp |
| **Kết nối nhiều PivotTable** | Phải lọc từng PivotTable riêng | 1 Slicer → lọc tất cả cùng lúc |
| **Phù hợp cho Dashboard** | ❌ Không chuyên nghiệp | ✅ Chuẩn mực Dashboard hiện đại |

---

### Cách Slicer hoạt động

```
User click nút "Miền Bắc" trên Slicer Region
            ↓
Slicer gửi filter tới TẤT CẢ PivotTables được kết nối
            ↓
    PT_Region     PT_Monthly     PT_Category     PT_Products
    (cập nhật)   (cập nhật)     (cập nhật)      (cập nhật)
            ↓
    Chart 1       Chart 2        Chart 3         Chart 4
    (tự cập nhật vì liên kết với PivotTable tương ứng)
```

Đây là điểm mạnh cốt lõi: **1 thao tác → toàn bộ Dashboard thay đổi theo**.

---

## 🎨 PHẦN 4 — FORMAT & STYLE SLICERS

---

### Chọn Slicer Style (màu sắc tổng thể)

Excel có sẵn bộ Slicer Styles — cách nhanh nhất để đồng bộ màu với Dashboard:

1. Click vào Slicer
2. Tab **Slicer** → nhóm **Slicer Styles**
3. Hover để preview → chọn style phù hợp với Color Palette của Dashboard

**Gợi ý chọn style:**
- Nếu Dashboard dùng tone xanh dương: chọn **"Slicer Style Dark 2"** (xanh đậm) hoặc **"Slicer Style Light 2"**
- Tránh các style quá sặc sỡ (đỏ, cam, tím) trừ khi chủ đích muốn nổi bật

---

### Tạo Custom Slicer Style

Nếu không có style vừa ý, tạo mới:

1. Tab **Slicer** → **Slicer Styles** → cuộn xuống → **New Slicer Style**
2. Trong hộp thoại, chọn từng **Slicer Element** rồi format:

| Element | Format gợi ý |
|---|---|
| **Whole Slicer** | Border: None; Font: Calibri 10 |
| **Header** | Fill: màu Primary (`#1D4ED8`); Font: White, Bold |
| **Selected Item with Data** | Fill: màu Primary (`#1D4ED8`); Font: White |
| **Unselected Item with Data** | Fill: White; Font: Dark grey |
| **No Data Item** | Fill: Light grey; Font: Grey (ô mờ = không có data) |

---

### Đồng bộ tất cả Slicers cùng style

Sau khi tạo custom style:
1. Ctrl+click để chọn cả 3 Slicers cùng lúc
2. Tab **Slicer** → **Slicer Styles** → chọn custom style vừa tạo
3. Cả 3 Slicer cùng áp dụng style — nhất quán về màu sắc

---

### Xóa Header Slicer (tùy chọn)

Mặc định Slicer có thanh Header hiển thị tên trường (ví dụ: "Region"). Đôi khi header thừa vì vị trí Slicer đã rõ mục đích:

1. Click vào Slicer → Tab **Slicer** → bỏ tick **"Display Header"**
2. Hoặc: Right-click → **Slicer Settings** → bỏ tick **"Display header"**

> **Khi nào nên giữ header:** Khi Slicer đặt xa vùng label hoặc người xem không rõ Slicer lọc theo tiêu chí gì.  
> **Khi nào ẩn header:** Khi đã có text box label riêng bên cạnh, hoặc khi muốn Slicer trông gọn hơn.

## 📊 PHẦN 5 — LÝ THUYẾT: KPI CARDS LÀ GÌ?

---

### KPI Card trong Dashboard

KPI (Key Performance Indicator) Card là các ô tóm tắt đặt ở vị trí nổi bật nhất trên Dashboard — thường ở hàng đầu tiên — hiển thị các con số quan trọng nhất mà ban lãnh đạo cần nhìn ngay khi mở file.

**Ví dụ KPI Cards trong Sales Dashboard:**

```
┌──────────────┐  ┌──────────────┐  ┌──────────────┐  ┌──────────────┐
│  DOANH THU   │  │  SỐ ĐƠN HÀNG │  │  TB / ĐƠN    │  │ LỢI NHUẬN    │
│              │  │              │  │              │  │              │
│   11.5 Tỷ    │  │    9,994     │  │   1,152K     │  │   3.2 Tỷ     │
└──────────────┘  └──────────────┘  └──────────────┘  └──────────────┘
```

---

### GETPIVOTDATA — Hàm đặc biệt cho KPI động

`GETPIVOTDATA` là hàm lấy dữ liệu từ PivotTable — **tự động cập nhật khi Slicer lọc**.

**Cú pháp cơ bản:**
```
=GETPIVOTDATA("tên_trường", ô_góc_pivottable)
```

**Ví dụ thực tế:**
```
=GETPIVOTDATA("Revenue", PT_Region!$A$3)
```

Giải thích:
- `"Revenue"` — tên trường muốn lấy (đúng tên trong PivotTable, phân biệt hoa thường)
- `PT_Region!$A$3` — địa chỉ ô góc trên bên trái của PivotTable trên sheet PT_Region

> 🎯 **Tại sao dùng GETPIVOTDATA thay vì tham chiếu ô trực tiếp?**  
> Nếu bạn gõ `=PT_Region!B5` để lấy Grand Total, khi PivotTable thay đổi cấu trúc (thêm dòng, sort lại), con số B5 có thể trỏ sai ô. `GETPIVOTDATA` luôn lấy đúng giá trị theo tên trường, bất kể PivotTable thay đổi như thế nào.

---

### Phân biệt KPI động vs KPI tĩnh

| Loại | Công thức | Khi Slicer lọc |
|---|---|---|
| **KPI động** | `=GETPIVOTDATA(...)` | ✅ Tự cập nhật theo Slicer |
| **KPI tĩnh** | `=SUM(SalesData[Revenue])` | ❌ Không thay đổi — luôn hiện toàn bộ data |

Trong Dashboard chuyên nghiệp, ưu tiên **KPI động** để người dùng thấy số liệu thay đổi khi họ lọc theo vùng/danh mục/quý.

---

### Progression kỹ năng KPI

| Cấp độ | Kỹ năng | Buổi này |
|---|---|---|
| 🟢 Cơ bản | Tạo ô KPI với SUM/COUNTA/AVERAGE | ✅ |
| 🟡 Trung cấp | Dùng GETPIVOTDATA để KPI tự cập nhật theo Slicer | ✅ |
| 🔵 Nâng cao | Thêm mũi tên tăng/giảm so với kỳ trước (IF + Conditional Formatting) | Tự học |
| 🔴 Chuyên nghiệp | KPI với Sparklines + Conditional Formatting động | Kỹ năng nâng cao |

## 🔢 PHẦN 6 — TẠO 4 KPI CARDS: CÔNG THỨC & CẤU TRÚC

---

### Vị trí đặt KPI Cards trên Dashboard

Trước khi nhập công thức, chuẩn bị vùng ô trên sheet `Dashboard`:

1. Hàng 1–3: Tiêu đề Dashboard (sẽ làm ở Buổi 6)
2. **Hàng 4–7: Vùng KPI Cards** — chia thành 4 cột đều nhau
3. Hàng 8 trở đi: Vùng Charts và Slicers

Merge cells cho mỗi KPI Card: chọn vùng khoảng **3 cột × 3 hàng** → Merge & Center. Tạo 4 vùng như vậy cạnh nhau.

---

### 4 KPI Cards: Công thức chi tiết

---

#### KPI Card 1 — Tổng Doanh Thu

| Thành phần | Nội dung |
|---|---|
| **Label** | `TỔNG DOANH THU` |
| **Công thức** | `=GETPIVOTDATA("Revenue",PT_Region!$A$3)` |
| **Format số** | `#,##0,,"Tỷ"` |
| **Ví dụ hiển thị** | `11 Tỷ` |

**Giải thích `#,##0,,"Tỷ"`:**
- `#,##0` — số có dấu phân cách hàng nghìn
- `,,` — chia cho 1,000,000 (2 dấu phẩy = chia 1 triệu)
- `"Tỷ"` — thêm chữ "Tỷ" sau số (1 tỷ VND ≈ 1,000,000,000 VND)

| **Ví dụ hiển thị** | `9,994` |

**Tại sao dùng `COUNTA` thay vì `COUNT`?**
- `COUNT` chỉ đếm ô có số
- `COUNTA` đếm tất cả ô không rỗng (kể cả text)
- `Order_ID` là mã chữ (ví dụ: "CA-2024-152156") → dùng `COUNTA`

> ⚠️ **Hạn chế:** `COUNTA(SalesData[Order_ID])` tính toàn bộ data, **không thay đổi khi lọc Slicer**. Nếu muốn KPI đếm đơn hàng theo Slicer, cần dùng GETPIVOTDATA với trường Count.

### Kỹ thuật nâng cao — KPI so sánh với kỳ trước

Thêm dòng nhỏ dưới Value để hiển thị `▲ +12%` hoặc `▼ -5%` so với cùng kỳ năm trước:

1. Thêm hàng thứ 3 vào KPI Card
2. Công thức: `=IF(năm_nay>năm_trước,"▲ ","▼ ")&TEXT(ABS((năm_nay-năm_trước)/năm_trước),"0%")`
3. Format: font nhỏ (9pt), màu xanh lá nếu tăng (`#16A34A`), màu đỏ nếu giảm (`#DC2626`)
4. Dùng **Conditional Formatting** để tự động đổi màu: `IF(value > 0, tô xanh, tô đỏ)`

Đây là kỹ thuật phổ biến trong báo cáo tài chính — chỉ cần làm được 4 KPI cơ bản trước, phần này có thể bổ sung sau.

### 💡 3 Takeaway quan trọng

**1. Report Connections = linh hồn của Interactive Dashboard**
Không có Report Connections, Slicer chỉ là filter trang trí. Với Report Connections đúng, 1 click của người dùng làm toàn bộ Dashboard sống động — đây là điểm khác biệt giữa báo cáo tĩnh và Dashboard thật sự.

**2. GETPIVOTDATA giúp KPI Cards "sống" cùng Slicer**
Dùng SUM/AVERAGE trực tiếp trên data → KPI không thay đổi khi lọc. Dùng GETPIVOTDATA → KPI tự cập nhật theo Slicer. Sự khác biệt nhỏ trong công thức, sự khác biệt lớn trong trải nghiệm người dùng.

**3. KPI Cards là điểm nhìn đầu tiên**
Trong thực tế, người xem Dashboard thường chỉ nhìn 10 giây đầu vào KPI Cards trước khi đọc Charts. 4 số quan trọng nhất phải nổi bật, đúng vị trí, đúng format. Buổi 6 sẽ hoàn thiện layout để KPI Cards luôn ở vị trí ưu tiên.
