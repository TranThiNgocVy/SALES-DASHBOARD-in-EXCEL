## 🤔 PHẦN 1 — PIVOTTABLE LÀ GÌ VÀ TẠI SAO CẦN DÙNG?

---

### PivotTable là gì?

**PivotTable** (Bảng tổng hợp xoay chiều) là công cụ phân tích dữ liệu mạnh nhất trong Excel — cho phép bạn **tóm tắt, nhóm, lọc và tính toán** hàng nghìn dòng dữ liệu chỉ bằng thao tác kéo thả, không cần viết một công thức nào.

> 💡 Tên "Pivot" (xoay) xuất phát từ khả năng nhìn dữ liệu từ nhiều góc độ khác nhau bằng cách thay đổi vị trí các trường phân tích.

---

### So sánh: Trước và sau khi có PivotTable

| Tình huống | Không có PivotTable | Có PivotTable |
|---|---|---|
| Tính tổng Revenue theo Region | Viết 5 công thức SUMIF | Kéo 2 trường, 5 giây |
| Phân tích theo tháng + Category | Pivot table thủ công, mất 1 giờ | 3 thao tác kéo thả |
| Filter Top 10 sản phẩm | Sort + cut thủ công | 1 click Value Filters |
| Thay đổi chiều phân tích | Viết lại toàn bộ | Kéo trường sang vị trí khác |
| Data mới thêm vào | Cập nhật lại tất cả công thức | 1 click Refresh |

---

### Khi nào dùng PivotTable?

PivotTable phù hợp khi bạn cần:
- **Tổng hợp nhanh** — "Tổng Revenue theo từng vùng miền là bao nhiêu?"
- **So sánh** — "Tháng nào bán tốt nhất? Category nào lợi nhuận cao nhất?"
- **Xếp hạng** — "Top 10 sản phẩm bán chạy nhất?"
- **Phân tích đa chiều** — "Revenue của Segment Corporate tại Region Miền Nam theo từng tháng?"
- **Làm nguồn dữ liệu** — PivotTable là **nguồn** để vẽ PivotChart trong các buổi tiếp theo

## 🖥️ PHẦN 2 — GIẢI PHẪU GIAO DIỆN PIVOTTABLE

---

### 4 Vùng kéo thả — Trái tim của PivotTable

Khi tạo PivotTable, bảng **PivotTable Fields** xuất hiện bên phải màn hình với 4 vùng:

| Vùng | Tên | Vai trò | Ví dụ với dataset |
|---|---|---|---|
| **FILTERS** | Bộ lọc tổng thể | Lọc toàn bộ bảng theo 1 tiêu chí | Lọc theo `Year` để chỉ xem 2025 |
| **COLUMNS** | Tiêu đề cột | Tạo thêm chiều ngang cho bảng | `Quarter` → tạo cột Q1, Q2, Q3, Q4 |
| **ROWS** | Tiêu đề hàng | Nhóm dữ liệu theo chiều dọc | `Region` → tạo 5 hàng vùng miền |
| **VALUES** | Giá trị tính toán | Số liệu cần tổng hợp (Sum, Count, Avg...) | `Revenue` → tính tổng Revenue từng vùng |

---

### Quy tắc vàng khi kéo thả

> 🎯 **Rows** = "Tôi muốn phân tích theo cái gì?" → kéo chiều phân tích vào đây  
> 🎯 **Values** = "Tôi muốn xem số liệu gì?" → kéo chỉ số cần đo lường vào đây  
> 🎯 **Filters** = "Tôi chỉ muốn xem trong điều kiện nào?" → kéo bộ lọc vào đây  
> 🎯 **Columns** = Dùng khi muốn so sánh song song (ít dùng hơn)

---

### Các hàm tính toán trong vùng VALUES

Khi kéo một cột số vào **Values**, mặc định Excel dùng `SUM`. Bạn có thể thay đổi:

| Hàm | Ý nghĩa | Khi nào dùng |
|---|---|---|
| `Sum` | Tổng | Tổng Revenue, Profit |
| `Count` | Đếm số dòng | Đếm số đơn hàng |
| `Average` | Trung bình | Doanh thu trung bình/đơn |
| `Max / Min` | Giá trị lớn nhất / nhỏ nhất | Đơn hàng lớn nhất |
| `% of Grand Total` | Tỷ lệ % so với tổng | Region chiếm bao nhiêu % tổng Revenue |

**Cách thay đổi hàm:**  
Click vào tên trường trong vùng **Values** → **Value Field Settings** → chọn hàm → OK

### Thêm cột Month_Name để hiển thị đẹp hơn

Hiện tại PivotTable chỉ hiển thị số tháng (1, 2, 3...). Để biểu đồ ở Buổi 4 hiển thị "Jan, Feb, Mar" thay vì số:

**Cách 1 — Dùng cả 2 trường:**
1. Kéo thêm `Month_Name` vào vùng **ROWS**, đặt bên dưới `Month`
2. PivotTable sẽ hiển thị phân cấp: số tháng → tên tháng

**Cách 2 — Chỉ dùng Month_Name (đơn giản hơn):**
1. Xóa `Month` khỏi Rows
2. Kéo `Month_Name` vào Rows thay thế
3. Nhược điểm: tháng sẽ sort theo alphabet (Apr, Aug, Dec...) thay vì thứ tự thực

> 💡 **Best practice:** Giữ cả `Month` (để sort đúng thứ tự) + `Month_Name` (để trục X của biểu đồ hiển thị đẹp). Trong PivotTable, dùng **Month** làm nền, còn khi vẽ chart sẽ dùng **Month_Name** làm label.

### 💡 3 Takeaway quan trọng

**1. PivotTable = SUMIF trên steroids**
Mọi thứ bạn làm được bằng SUMIF/COUNTIF đều làm được bằng PivotTable — nhanh hơn, linh hoạt hơn, và không lỗi khi thêm dữ liệu mới.

**2. Mỗi PivotTable = 1 góc nhìn về dữ liệu**
Không có "PivotTable đúng duy nhất". 6 PivotTable trong buổi này là 6 cách nhìn khác nhau vào cùng 1 dataset. Thực tế, một Dashboard thường cần 8–12 PivotTable.

**3. PT là nguồn — Chart là kết quả**
PivotTable không phải sản phẩm cuối cùng. Ở Buổi 4, mỗi PivotTable sẽ trở thành nguồn cho 1 biểu đồ trực quan trên Dashboard.
