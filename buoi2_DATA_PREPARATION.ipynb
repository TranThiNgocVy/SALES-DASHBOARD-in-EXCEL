## ➕ PHẦN 3 — TỰ TAY TẠO CÁC CỘT MỚI BẰNG CÔNG THỨC

---

> 📌 **Tại sao phải tự tạo?** Dữ liệu thô từ hệ thống thường chỉ xuất ra các trường cơ bản. Nhiệm vụ của Data Analyst là **biến đổi** dữ liệu thô thành các chỉ số có ý nghĩa phân tích. Đây là kỹ năng cốt lõi trong thực tế.

Giả sử dữ liệu bắt đầu từ **hàng 2**, `Order_Date` ở cột **B**, `Revenue` ở cột **Q**, `Cost` ở cột **R**  
*(Điều chỉnh tên cột theo thực tế file của bạn)*

---

### 💰 Cột 1 — `Profit` (Lợi nhuận)

**Ý nghĩa:** Số tiền thực lãi sau khi trừ giá vốn. Có thể **âm** nếu đơn bị chiết khấu quá cao.

**Công thức** (nhập vào cột trống đầu tiên sau cột Cost, ví dụ cột S2):
```excel
= Q2 - R2
```
*(Revenue − Cost)*

**Kết quả mẫu:** `23,950,640` — sau khi format sẽ hiển thị là `23,950,640 VND`


![img2-2](images/images_buoi2/img2-2.png)
---

### 📈 Cột 2 — `Profit_Margin` (Tỷ suất lợi nhuận)

**Ý nghĩa:** Phần trăm lợi nhuận trên mỗi đồng doanh thu — chỉ số quan trọng nhất để so sánh hiệu quả giữa các sản phẩm, danh mục.

**Công thức** (cột T2):
```excel
= S2 / Q2
```
*(Profit ÷ Revenue)*

**Kết quả mẫu:** `0.38` — sau khi format sẽ hiển thị là `38%`


![img2-3](images/images_buoi2/img2-3.png)
> ⚠️ Nếu ô Profit hoặc Revenue = 0, công thức sẽ báo lỗi `#DIV/0!`. Dùng công thức an toàn hơn:
> ```excel
> = IF(Q2=0, 0, S2/Q2)
> ```

---

### 📅 Cột 3 — Nhóm cột thời gian (Month, Month_Name, Quarter, Year, Month_Year)

**Ý nghĩa:** Tách `Order_Date` thành các chiều thời gian riêng lẻ để dễ phân tích theo tháng, quý, năm trong PivotTable.

| Cột | Công thức | Kết quả mẫu | Ghi chú |
|---|---|---|---|
| `Month` | `=MONTH(B2)` | `7` | Số tháng (1–12) |
| `Month_Name` | `=TEXT(B2,"mmm")` | `Jul` | Tên tháng viết tắt tiếng Anh |
| `Quarter` | `="Q"&INT((MONTH(B2)-1)/3+1)` | `Q3` | Tên quý Q1–Q4 |
| `Year` | `=YEAR(B2)` | `2024` | Năm 4 chữ số |
| `Month_Year` | `=TEXT(B2,"mmm-yyyy")` | `Jul-2024` | Nhãn kết hợp tháng-năm cho biểu đồ |

> 💡 **Tại sao cần cả 5 cột?** Mỗi cột phục vụ một mục đích khác nhau trong PivotTable và Chart:
> - `Month` (số): dùng để **sort** đúng thứ tự tháng
> - `Month_Name`: dùng làm **nhãn trục X** trên biểu đồ
> - `Quarter`: dùng **filter theo quý**
> - `Year`: dùng **so sánh năm này với năm trước**
> - `Month_Year`: dùng làm **nhãn khi cần hiển thị cả tháng lẫn năm** trên biểu đồ

### 🏷️ Cột 4 — `Promotion` (Chương trình khuyến mãi)

**Ý nghĩa:** Xác định đơn hàng nào được áp dụng khuyến mãi và thuộc loại khuyến mãi nào.

**Công thức gợi ý**:

*(Nếu Discount = 0 → không có KM; nếu có → "Sale " + tên Category)*

### ⚠️ Lưu ý đặc biệt cho cột Year

Nếu bạn format cột `Year` theo dạng `Number` mặc định, Excel sẽ hiển thị `2,024` (có dấu phẩy phân cách hàng nghìn) — **trông rất kỳ lạ** khi làm nhãn trục X trên biểu đồ.

### 🟢 Bài 1 (Cơ bản) — Đọc hiểu dataset

Mở file `SalesDataSetup.xlsx` và dùng phím tắt `Ctrl + End` để tìm ô cuối cùng.

**Câu hỏi:** Dataset có bao nhiêu dòng dữ liệu (không tính header)? Có bao nhiêu cột trước khi thêm cột mới?

**✅ Đáp án:** Nhìn vào ô cuối cùng — số dòng là số hàng của ô đó trừ 1 (vì dòng 1 là header). Số cột là ký tự cột tại ô đó (ví dụ cột R = 18 cột).

---

### 🟢 Bài 2 (Cơ bản) — Nhập công thức Profit

Tạo cột `Profit` ngay sau cột `Cost`. Nhập công thức cho dòng 2 và fill xuống toàn bộ.

**✅ Đáp án:**
```excel
= [ô Sales dòng 2] - [ô Cost dòng 2]
```
Ví dụ nếu Sales ở Q2 và Cost ở R2:
```excel
= Q2 - R2
```
Double-click góc dưới phải để fill xuống hết.

---

### 🟢 Bài 3 (Cơ bản) — Format đơn vị

Sau khi tạo cột `Profit`, các số đang hiển thị dạng thô (ví dụ `23950640`). Hãy format để hiển thị đẹp như: `23,950,640 VND`.

**✅ Đáp án:**
1. Chọn toàn bộ cột Profit (click vào tên cột ở header)
2. Nhấn `Ctrl + 1` → tab **Number** → chọn **Custom**
3. Nhập vào ô Type: `#,##0 "VND"`
4. Click OK

---

### 🟡 Bài 4 (Trung cấp) — Tạo cột Quarter

Tạo cột `Quarter` từ cột `Order_Date` sao cho kết quả hiển thị là `Q1`, `Q2`, `Q3`, `Q4`.

**✅ Đáp án:**
```excel
= "Q" & INT((MONTH(B2)-1)/3+1)
```
- `MONTH(B2)` lấy số tháng (1–12)
- `INT((x-1)/3+1)` chuyển tháng thành số quý (1–4)

### 💡 3 Takeaway quan trọng

**1. Hiểu dữ liệu trước khi phân tích**
Data Analyst không bao giờ tạo biểu đồ ngay khi nhận data. Bước đầu tiên luôn là **đọc kỹ cấu trúc** — hiểu từng cột nói về gì, kiểu dữ liệu là gì, phạm vi giá trị như thế nào.

**2. Cột tính toán = giá trị gia tăng**
`Profit`, `Profit_Margin`, `Quarter`... không có trong data thô nhưng lại là những chỉ số quan trọng nhất trên Dashboard. Khả năng tạo ra các cột phái sinh này phân biệt người dùng Excel phổ thông với Data Analyst.

**3. Excel Table = nền tảng cho Dashboard tự cập nhật**
Nếu bỏ bước này, Dashboard sẽ không tự cập nhật khi có dữ liệu mới. Đây là lý do nhiều file Excel trong thực tế bị "chết" — phải cập nhật tay mỗi tháng.
