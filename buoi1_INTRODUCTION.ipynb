## 🤔 PHẦN 1 — TẠI SAO CẦN SALES DASHBOARD?

---

### 🧩 Vấn đề: Bảng số liệu không ai đọc được

Hãy tưởng tượng bạn là **Sales Director** của một công ty. Mỗi tuần, bộ phận kế toán gửi cho bạn một file Excel có **5.000 dòng** dữ liệu bán hàng như thế này:

```
Ngày        | Mã SP  | Tên SP         | Vùng   | Doanh thu  | Số lượng
------------|--------|----------------|--------|------------|----------
01/01/2024  | SP001  | Áo thun nam    | HCM    | 250,000    | 5
01/01/2024  | SP002  | Quần jeans     | HN     | 450,000    | 2
01/01/2024  | SP003  | Giày sneaker   | ĐN     | 800,000    | 1
...         | ...    | ...            | ...    | ...        | ...
(5000 dòng tiếp theo)
```

**Câu hỏi:** Từ bảng dữ liệu trên, bạn có thể trả lời ngay được không?
- 🔴 Tháng này doanh thu tăng hay giảm so với tháng trước?
- 🔴 Vùng nào đang bán tốt nhất?
- 🔴 Sản phẩm nào cần đẩy mạnh?

> **Câu trả lời thực tế:** Không! Bạn phải mất **2-4 tiếng** ngồi lọc, tính toán, tổng hợp mới ra được báo cáo!

### 🔧 Kỹ năng Excel — Lộ trình từ Cơ bản đến Chuyên nghiệp

> Mỗi kỹ năng được xây dựng **có hệ thống qua 7 buổi** — từ hiểu khái niệm, thực hành cơ bản, đến ứng dụng chuyên nghiệp trong Dashboard thực tế.

| Kỹ năng | Cơ bản 🟢 | Trung cấp 🟡 | Nâng cao 🔵 | Chuyên nghiệp 🔴 |
|---|---|---|---|---|
| 📋 **Excel Tables** | Tạo Table bằng `Ctrl+T` | Đặt tên Table, hiểu Structured References | Table tự mở rộng khi thêm data | Dùng Table làm nguồn cho PivotTable & Chart |
| 🔄 **PivotTables** | Tạo PivotTable đơn giản | Thêm/xóa Fields, thay đổi phép tính | Grouping theo tháng/quý, Calculated Fields | Tối ưu PivotTable làm nguồn cho toàn bộ Dashboard |
| 📈 **Charts** | Tạo chart từ PivotTable | Chỉnh title, màu, trục | Kết hợp nhiều series, Secondary Axis | Chart tự động cập nhật, đồng bộ với Slicer |
| 🎛️ **Slicers** | Tạo Slicer cho 1 PivotTable | Kết nối Slicer với nhiều PivotTable | Tùy chỉnh style Slicer theo màu Dashboard | Slicer điều khiển toàn bộ Dashboard cùng lúc |
| 🃏 **KPI Cards** | Hiểu KPI Card là gì | Dùng `GETPIVOTDATA` lấy số từ PivotTable | Format số: tiền tệ, %, làm tròn | Card tự động đổi màu theo ngưỡng (Conditional Formatting) |
| 🎨 **Formatting** | Tô màu, in đậm, căn chỉnh | Conditional Formatting theo điều kiện | Xây dựng color theme nhất quán | Dashboard "pixel-perfect" — chuyên nghiệp như Power BI |
| 🔢 **Hàm hỗ trợ** | `SUM`, `COUNT`, `AVERAGE` | `GETPIVOTDATA`, `IFERROR` | `TEXT`, `IF`, `SUMIFS` | Kết hợp hàm để tạo KPI động, tự cập nhật |

## 📝 PHẦN 5 — BÀI TẬP THỰC HÀNH BUỔI 1

> 💡 **Hướng dẫn:** Các bài tập này là **câu hỏi tư duy + thực hành nhỏ** — không cần mở Excel ngay hôm nay. Hãy **tự suy nghĩ trước**, sau đó đọc gợi ý đáp án bên dưới để so sánh với câu trả lời của mình.

---

### 🟢 Bài 1 — Xác định vấn đề *(Cơ bản)*

**Tình huống:** Bạn là nhân viên bán hàng tại một shop thời trang có 3 cửa hàng ở Hà Nội, Đà Nẵng, TP.HCM. Mỗi tháng bạn nhận được 3 file Excel riêng biệt từ mỗi cửa hàng.

**Câu hỏi:**
1. Liệt kê ít nhất **3 vấn đề** bạn gặp phải khi tổng hợp báo cáo từ 3 file riêng lẻ.
2. Nếu có một Dashboard, những vấn đề đó được giải quyết như thế nào?

**💡 Gợi ý đáp án:**

**Vấn đề khi không có Dashboard:**

| # | Vấn đề | Mô tả |
|---|---|---|
| 1 | ⏰ **Tốn thời gian** | Phải mở 3 file, copy-paste thủ công vào file tổng hợp |
| 2 | ❌ **Dễ sai sót** | Copy nhầm dòng, paste sai vị trí → số liệu sai |
| 3 | 😵 **Khó so sánh** | 3 file format khác nhau, khó nhìn tổng thể |
| 4 | 🔄 **Phải làm lại mỗi tháng** | Không tự động, lặp đi lặp lại vô tận |
| 5 | 👥 **Không tự phục vụ** | Manager phải chờ nhân viên tổng hợp xong |

**Dashboard giải quyết:**
- 📁 Tất cả data từ 3 cửa hàng vào **1 file duy nhất** → không cần mở nhiều file
- 🔄 PivotTable tự tổng hợp → **không copy-paste → không sai số**
- 🎛️ Slicer lọc theo cửa hàng → **so sánh dễ dàng trong 1 click**
- ⚡ Khi data mới thêm vào, dashboard tự cập nhật → **tiết kiệm 100% thời gian**

---

### 🟢 Bài 2 — Phân loại KPI *(Cơ bản)*

**Tình huống:** Bạn là Sales Analyst được yêu cầu xây dashboard cho bộ phận bán hàng.

**Câu hỏi:** Sắp xếp các chỉ số sau vào đúng nhóm: **KPI Card**, **Biểu đồ**, **Thông tin phụ**.

```
Danh sách chỉ số cần phân loại:
  • Tổng doanh thu tháng          • Doanh thu theo từng ngày
  • Tên nhân viên bán hàng        • Số đơn hàng
  • Tỷ lệ hủy đơn theo sản phẩm  • Lợi nhuận gộp
  • Doanh thu theo vùng miền      • Địa chỉ kho hàng
  • Giá trị đơn hàng trung bình (AOV)
```

**💡 Gợi ý đáp án:**

| Nhóm | Chỉ số thuộc nhóm |
|---|---|
| 🏆 **KPI Card** *(số to, nổi bật)* | Tổng doanh thu, Số đơn hàng, Lợi nhuận gộp, AOV |
| 📈 **Biểu đồ** *(trực quan hóa)* | Doanh thu theo ngày → Line Chart; Tỷ lệ hủy đơn → Bar Chart; Doanh thu theo vùng → Column Chart |
| ℹ️ **Thông tin phụ** *(không cần trên Dashboard)* | Tên nhân viên, Địa chỉ kho hàng |

> 🎯 **Nguyên tắc:** Dashboard chỉ chứa **những gì giúp ra quyết định**. Thông tin chi tiết (tên nhân viên, địa chỉ) để nguyên ở file data gốc.

### 🟡 Bài 3 — Thiết kế Layout *(Trung bình)*

**Tình huống:** Bạn được giao thiết kế bố cục Dashboard bán hàng với các phần tử sau:
- 4 KPI Cards (Revenue, Orders, Profit, AOV)
- 1 Line Chart (doanh thu theo tháng)
- 1 Bar Chart (top sản phẩm)
- 1 Doughnut Chart (tỷ trọng theo danh mục)
- 3 Slicers (Vùng, Danh mục, Quý)
- Tiêu đề Dashboard

**Câu hỏi:** Hãy mô tả bằng chữ hoặc vẽ sơ đồ cách bạn sắp xếp các phần tử này. Bạn sẽ đặt gì **ở hàng đầu tiên**? Cái gì quan trọng nhất để **ở đâu**?

**💡 Gợi ý đáp án — Bố cục phổ biến và hiệu quả:**

```
┌──────────────────────────────────────────────────────────────┐
│         Tiêu đề: SALES DASHBOARD 2024                        │
├──────────────┬──────────────────┬───────────────────────────┤
│ Slicer: Vùng │ Slicer: Danh mục │ Slicer: Quý               │
├──────────────┴──────────────────┴───────────────────────────┤
│  KPI: Revenue │ KPI: Orders │ KPI: Profit │ KPI: AOV         │
├──────────────────────────────────────────────────────────────┤
│          Line Chart: Doanh thu theo tháng (rộng)             │
├─────────────────────────────────┬────────────────────────────┤
│ Bar Chart: Top sản phẩm bán chạy│ Doughnut: Tỷ trọng DM     │
└─────────────────────────────────┴────────────────────────────┘
```

**Nguyên tắc bố cục Dashboard:**

| Vị trí | Đặt gì | Lý do |
|---|---|---|
| ⬆️ **Trên cùng** | Tiêu đề + Slicers | Người xem biết ngay đang xem gì và có thể filter |
| 🔢 **Giữa trên** | KPI Cards | Số quan trọng nhất cần thấy đầu tiên |
| 📊 **Phần dưới** | Charts | Chi tiết & phân tích — xem sau khi đã nắm KPI |

> 👁️ **F-pattern:** Mắt người đọc thường quét từ **trái → phải**, **trên → dưới** — đặt thông tin quan trọng ở góc trên trái!

**Nếu không có Dashboard:**
- ❌ Không thể biết trong 30 giây — phải lọc, group, tính toán thủ công
- 🕐 Biết chi nhánh nào tăng/giảm → cần SUMIF hoặc PivotTable thủ công
- 🕑 Biết danh mục nào tăng → phải làm thêm bảng tổng hợp riêng
- ⏱️ **Tổng thời gian: 30–60 phút** thay vì 30 giây

### 🟡 Bài 5 — Tư duy dữ liệu *(Trung bình - Khá)*

**Tình huống:** Bạn làm việc tại một cửa hàng đồ gia dụng. Bạn có dữ liệu bán hàng 12 tháng gần nhất.

**Câu hỏi:**
1. Bạn muốn biết **3 câu hỏi kinh doanh** quan trọng nhất từ dữ liệu đó?
2. Mỗi câu hỏi cần **loại biểu đồ / KPI nào** để trả lời?
3. **Ai** sẽ xem Dashboard này và họ cần thông tin gì nhất?

> 💭 **Không có đáp án duy nhất đúng.** Đây là bài tập tư duy — mỗi người sẽ có góc nhìn khác nhau!

**💡 Ví dụ đáp án tham khảo:**

**3 câu hỏi kinh doanh và biểu đồ phù hợp:**

| # | Câu hỏi kinh doanh | Loại biểu đồ / KPI |
|---|---|---|
| 1 | 💰 Tháng nào doanh thu cao nhất / thấp nhất? | **Line Chart** theo tháng + **KPI Card** doanh thu |
| 2 | 🏆 Top 5 sản phẩm bán chạy nhất? | **Bar Chart** sắp xếp theo doanh số |
| 3 | 📦 Danh mục nào đóng góp nhiều nhất vào doanh thu? | **Doughnut Chart** tỷ trọng theo danh mục |

**Đối tượng xem và nhu cầu:**

| Đối tượng | Nhu cầu chính |
|---|---|
| 👔 **Giám đốc** | KPI tổng quan, xu hướng nhanh — không cần quá chi tiết |
| 📊 **Analyst** | Filter nhiều chiều, số chi tiết, drill-down được |
| 🏪 **Quản lý cửa hàng** | Chỉ cần số của cửa hàng mình, so sánh với target |

> 🎯 **Bài học quan trọng:** Dashboard phải được thiết kế **phù hợp với người xem cụ thể**, không phải nhồi nhét mọi thứ vào một trang!
